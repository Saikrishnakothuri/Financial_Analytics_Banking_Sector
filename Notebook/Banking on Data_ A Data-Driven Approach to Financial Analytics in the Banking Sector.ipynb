```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "A Data-Driven Approach to Financial Analytics in the Banking Sector"
data = pd.read_excel("/content/drive/MyDrive/My Data/A Data-Driven Approach to Financial Analytics in the Banking Sector.xlsx")

# Display basic info
print(data.info())
print(data.head())
print(data.shape())

# Gender distribution
gender_counts = data["Gender"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title("Gender Distribution")
plt.axis("equal")
plt.show()

# Age Group Distribution
age_counts = data["Age"].value_counts(normalize=True) * 100
sns.barplot(x=age_counts.index, y=age_counts.values, palette="pastel")
plt.title("Age Group Distribution")
plt.ylabel("Percentage")
plt.xlabel("Age Groups")
plt.show()

#Frequency of Digital Banking Usage
usage_counts = data["2. How often do you use digital banking services (e.g., mobile banking, UPI, online transactions)?"].value_counts()
usage_counts.plot(kind="barh", color='teal', edgecolor='black')
plt.title("Frequency of Digital Banking Usage")
plt.xlabel("Count")
plt.ylabel("Usage Frequency")
plt.show()

#Primary Concerns About Banks Using Data
sns.countplot(y="5. What is your primary concern regarding banks using your financial data?", data=data, palette="viridis")
plt.title("Primary Concerns About Data Usage")
plt.show()


# Primary Concerns About Banks Using Data
#Corrected column name to match the actual column header
concern_column = "5. What is your primary concern regarding banks using your financial data?"
sns.countplot(y=concern_column, data=data, palette="viridis")
plt.title("Primary Concerns About Data Usage")
plt.show()

# Confidence in Fraud Detection Using AI
#Corrected column name to match the actual column header
confidence_column = "7. How confident are you in your bankâ€™s ability to detect and prevent fraudulent transactions using AI?"
sns.boxplot(x=confidence_column, data=data, palette="cool")
plt.title("Confidence in Fraud Detection Using AI")
plt.xticks(rotation=45, ha="right") # Rotate x-axis labels for better readability
plt.show()

#Impact of AI on Banking Experience
ai_impact = data["9. Do you believe AI-powered banking has improved your overall banking experience?"].value_counts()
plt.pie(ai_impact, labels=ai_impact.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("coolwarm", len(ai_impact)))
plt.title("Impact of AI on Banking Experience")
plt.axis("equal")
plt.show()

#AI Impact by Gender
sns.countplot(x="9. Do you believe AI-powered banking has improved your overall banking experience?", hue="Gender", data=data, palette="husl")
plt.title("AI Banking Experience Impact by Gender")
plt.legend(title="Gender")
plt.xticks(rotation=45)
plt.show()

#AI Impact Across Age Groups
sns.countplot(x="9. Do you believe AI-powered banking has improved your overall banking experience?", hue="Age", data=data, palette="pastel")
plt.title("AI Banking Experience Impact Across Age Groups")
plt.legend(title="Age Group")
plt.xticks(rotation=45)
plt.show()

#Security Features vs Trust in Banks
sns.countplot(y="8. What additional security features would make you feel safer while using digital banking?", hue="12. Do you trust banks to handle your financial data responsibly?", data=data, palette="muted")
plt.title("Security Feature Preferences vs Trust in Banks")
plt.legend(title="Trust Level")
plt.show()

#Emerging Banking Technology Trends
technology_trends = data["16. Which emerging banking technology do you think will have the biggest impact?"].value_counts()
technology_trends.plot(kind='line', color='green', marker='o')
plt.title("Emerging Banking Technology Trends")
plt.xlabel("Technologies")
plt.ylabel("Respondent Count")
plt.xticks(rotation=45)
plt.show()

#Relation Between Security Concerns and AI Impact
sns.scatterplot(x="5. What is your primary concern regarding banks using your financial data?", y="9. Do you believe AI-powered banking has improved your overall banking experience?", data=data, hue="Gender", palette="Set2")
plt.title("Security Concerns vs AI Impact on Experience")
plt.xticks(rotation=45)
plt.show()

#Confidence in Fraud Detection by AI
sns.boxplot(x="7. How confident are you in your bankâ€™s ability to detect and prevent fraudulent transactions using AI?", data=data, palette="cool")
plt.title("Confidence in Fraud Detection Using AI")
plt.show()

#Trust in Banks Handling Financial Data Responsibly vs Willingness to Share Data
sns.countplot(x="12. Do you trust banks to handle your financial data responsibly?", hue="10. Would you be willing to share more of your financial data if it meant better banking recommendations and security?", data=data, palette="pastel")
plt.title("Trust in Banks vs Willingness to Share Data")
plt.xticks(rotation=45)
plt.show()

#AI Awareness and Fraud Alerts
sns.countplot(x="3. Are you aware that banks use data analytics for fraud detection, risk assessment, and customer service?", hue="6. Have you ever received a fraud alert from your bank (e.g., unauthorized transaction notifications)?", data=data, palette="rocket")
plt.title("AI Awareness vs Fraud Alert Experience")
plt.xticks(rotation=45)
plt.show()

#Impact of AI-Powered Banking on Customer Experience
sns.countplot(x="9. Do you believe AI-powered banking has improved your overall banking experience?", hue="7. How confident are you in your bankâ€™s ability to detect and prevent fraudulent transactions using AI?", data=data, palette="husl")
plt.title("AI-Powered Banking Experience vs Confidence in AI")
plt.xticks(rotation=45)
plt.show()

#Digital Banking Frequency and AI Awareness
sns.countplot(x="2. How often do you use digital banking services (e.g., mobile banking, UPI, online transactions)?", hue="3. Are you aware that banks use data analytics for fraud detection, risk assessment, and customer service?", data=data, palette="viridis")
plt.title("Digital Banking Frequency vs AI Awareness")
plt.xticks(rotation=45)
plt.legend(loc="upper right", bbox_to_anchor=(1.5, 1))
plt.show()

#Customer Feedback on Transparency Preferences
sns.countplot(x="14. Would you prefer banks to be more transparent about how they use your data for analytics?", data=data, palette="crest")
plt.title("Preferences for Data Transparency")
plt.show()

#Improvement Suggestions for Digital Banking Services
sns.countplot(y="18. What improvements would you like to see in digital banking services?", data=data, palette="dark")
plt.title("Suggestions for Improvement in Digital Banking")
plt.show()

```
